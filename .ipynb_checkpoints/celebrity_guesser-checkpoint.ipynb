# Celebrity Guesser CNN

A Convolutional Neural Network trained on celebrity face images that can recognize 17 different celebrities.
The model is also tested on **partial/cropped images** (e.g. only eyes or half-face) to demonstrate generalization limitations.

**Dataset:** 17 celebrities × 100 images each = 1,700 images

**Classes:** Angelina Jolie, Brad Pitt, Denzel Washington, Hugh Jackman, Jennifer Lawrence, Johnny Depp, Kate Winslet, Leonardo DiCaprio, Megan Fox, Natalie Portman, Nicole Kidman, Robert Downey Jr., Sandra Bullock, Scarlett Johansson, Tom Cruise, Tom Hanks, Will Smith

```python
# Imports
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image as keras_image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import warnings
warnings.filterwarnings('ignore')

print(f"TensorFlow version: {tf.__version__}")
```

## 1. Data Loading & Preprocessing

```python
# Configuration
DATA_DIR = 'datasets/celeb'
IMG_SIZE = 64       # Resize all images to 64x64
BATCH_SIZE = 32
EPOCHS = 30
RANDOM_SEED = 42

# Get class names from folder names
CLASS_NAMES = sorted([d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d))])
NUM_CLASSES = len(CLASS_NAMES)

print(f"Found {NUM_CLASSES} classes:")
for i, name in enumerate(CLASS_NAMES):
    print(f"  {i}: {name}")
print(f"\nImage size: {IMG_SIZE}x{IMG_SIZE}")
```

```python
# Count total images
total_images = 0
for class_name in CLASS_NAMES:
    class_path = os.path.join(DATA_DIR, class_name)
    n = len(os.listdir(class_path))
    total_images += n
    print(f"{class_name}: {n} images")

print(f"\nTotal: {total_images} images")
```

```python
# ImageDataGenerator for data augmentation and preprocessing
# Only use augmentation for training data
train_datagen = ImageDataGenerator(
    rescale=1./255,              # Normalize pixel values to [0, 1]
    rotation_range=15,           # Random rotation up to 15 degrees
    width_shift_range=0.1,      # Random horizontal shift
    height_shift_range=0.1,     # Random vertical shift
    horizontal_flip=True,       # Random horizontal flip
    zoom_range=0.1,             # Random zoom
    validation_split=0.2        # Hold out 20% for validation
)

# For test data, only rescale (no augmentation)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create training generator (80% of data)
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    seed=RANDOM_SEED
)

# Create validation generator (20% of data)
val_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    seed=RANDOM_SEED
)

print(f"\nTraining samples: {train_generator.samples}")
print(f"Validation samples: {val_generator.samples}")
print(f"\nClass indices: {train_generator.class_indices}")
```

```python
# Visualize sample images from training set
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

for i in range(10):
    img, label = train_generator[i % len(train_generator)][0][0], train_generator[i % len(train_generator)][1][0]
    class_idx = np.argmax(label)
    axes[i].imshow(img)
    axes[i].set_title(CLASS_NAMES[class_idx], fontsize=9)
    axes[i].axis('off')

plt.suptitle('Sample Celebrity Images', fontsize=14)
plt.tight_layout()
plt.show()
```

## 2. CNN Model Architecture

```python
def build_cnn_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=NUM_CLASSES):
    """
    Build a CNN model for celebrity classification.
    
    Architecture:
    - 3 Convolutional blocks (Conv2D + MaxPooling)
    - Increasing filter sizes: 32 -> 64 -> 128
    - Dropout for regularization
    - Dense layers for classification
    """
    model = keras.Sequential([
        # Block 1: 32 filters
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 2: 64 filters
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 3: 128 filters
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Classification head
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

model = build_cnn_model()
model.summary()
```

```python
# Compile the model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

## 3. Training

```python
# Callbacks
callbacks = [
    # Stop early if validation loss doesn't improve
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    ),
    # Reduce learning rate when plateau
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-6
    ),
    # Save best model
    keras.callbacks.ModelCheckpoint(
        'best_celeb_model.keras',
        monitor='val_accuracy',
        save_best_only=True
    )
]

print("Starting training...")
```

```python
# Train the model
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks
)
```

```python
# Plot training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Accuracy plot
axes[0].plot(history.history['accuracy'], label='Train Accuracy', marker='o')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy', marker='s')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Loss plot
axes[1].plot(history.history['loss'], label='Train Loss', marker='o')
axes[1].plot(history.history['val_loss'], label='Val Loss', marker='s')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print best results
best_val_acc = max(history.history['val_accuracy'])
best_epoch = np.argmax(history.history['val_accuracy']) + 1
print(f"\nBest validation accuracy: {best_val_acc:.4f} at epoch {best_epoch}")
```

## 4. Evaluation on Full Images

```python
# Load the best saved model
model = keras.models.load_model('best_celeb_model.keras')
print("Loaded best model from disk.")
```

```python
# Evaluate on validation set
val_loss, val_acc = model.evaluate(val_generator)
print(f"\nValidation Accuracy: {val_acc:.4f} ({val_acc*100:.2f}%)")
print(f"Validation Loss: {val_loss:.4f}")
```

```python
# Get predictions on validation set
val_generator.reset()  # Reset to start from beginning
Y_pred = model.predict(val_generator, verbose=0)
y_pred = np.argmax(Y_pred, axis=1)
y_true = val_generator.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASS_NAMES)

fig, ax = plt.subplots(figsize=(14, 12))
disp.plot(ax=ax, cmap='Blues', xticks_rotation=45, values_format='d')
plt.title('Confusion Matrix - Full Image Predictions', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES))
```

## 5. Test on Partial/Cropped Images

This section tests the model's ability to recognize celebrities from **partial face images** (cropped eyes, half-face, etc.).
This demonstrates the model's limitations when trained only on full faces — it typically fails on cropped inputs.

```python
def load_and_preprocess_image(img_path, target_size=(IMG_SIZE, IMG_SIZE)):
    """Load and preprocess a single image."""
    img = keras_image.load_img(img_path, target_size=target_size)
    img_array = keras_image.img_to_array(img)
    img_array = img_array / 255.0  # Normalize to [0, 1]
    return img_array


def crop_image(img, crop_type='eyes'):
    """
    Crop an image to simulate partial face inputs.
    
    crop_type options:
    - 'eyes': Top third (eye region only)
    - 'lower': Bottom two-thirds (mouth/chin region)
    - 'left_half': Left half of face
    - 'right_half': Right half of face
    - 'center': Central crop (60%)
    """
    h, w = img.shape[:2]
    
    if crop_type == 'eyes':
        # Top third (eye area)
        start_h = 0
        end_h = h // 3
    elif crop_type == 'lower':
        # Bottom two-thirds
        start_h = h // 4
        end_h = h
    elif crop_type == 'left_half':
        start_h = 0
        end_h = h
        w = w // 2
    elif crop_type == 'right_half':
        start_h = 0
        end_h = h
        w = w // 2
        start_w = w
    elif crop_type == 'center':
        # 60% center crop
        crop_h = int(h * 0.6)
        crop_w = int(w * 0.6)
        start_h = (h - crop_h) // 2
        end_h = start_h + crop_h
        start_w = (w - crop_w) // 2
        end_w = start_w + crop_w
        return img[start_h:end_h, start_w:end_w]
    else:
        return img
    
    if crop_type == 'left_half':
        return img[start_h:end_h, :w]
    elif crop_type == 'right_half':
        return img[start_h:end_h, start_w:]
    else:
        return img[start_h:end_h, :]


def predict_on_cropped_images(model, data_dir, class_names, crop_types=['eyes', 'lower', 'left_half', 'right_half', 'center']):
    """
    Test the model on cropped/partial images for each celebrity.
    
    Returns a dictionary with results per crop type.
    """
    results = {}
    
    for crop_type in crop_types:
        print(f"\n--- Testing with '{crop_type}' crop ---")
        correct = 0
        total = 0
        details = []
        
        for class_idx, class_name in enumerate(class_names):
            class_path = os.path.join(data_dir, class_name)
            images = os.listdir(class_path)[:10]  # Test on 10 images per class
            
            for img_name in images:
                img_path = os.path.join(class_path, img_name)
                
                # Load image
                img = load_and_preprocess_image(img_path)
                
                # Crop
                cropped = crop_image(img, crop_type)
                
                # Resize cropped back to model input size
                cropped_resized = tf.image.resize(cropped, (IMG_SIZE, IMG_SIZE))
                
                # Add batch dimension
                input_array = np.expand_dims(cropped_resized.numpy(), axis=0)
                
                # Predict
                probs = model.predict(input_array, verbose=0)[0]
                pred_idx = np.argmax(probs)
                pred_name = class_names[pred_idx]
                confidence = probs[pred_idx]
                
                is_correct = (pred_idx == class_idx)
                if is_correct:
                    correct += 1
                total += 1
                
                details.append({
                    'true': class_name,
                    'pred': pred_name,
                    'confidence': confidence,
                    'correct': is_correct
                })
        
        acc = correct / total
        results[crop_type] = {
            'accuracy': acc,
            'correct': correct,
            'total': total,
            'details': details
        }
        print(f"Accuracy: {acc:.4f} ({correct}/{total})")
    
    return results


# Run the cropped image test
crop_results = predict_on_cropped_images(model, DATA_DIR, CLASS_NAMES)
```

```python
# Summary of partial image results
crop_types = list(crop_results.keys())
accuracies = [crop_results[ct]['accuracy'] for ct in crop_types]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(crop_types, accuracies, color=['#3498db', '#e74c3c', '#2ecc71', '#9b59b6', '#f39c12'])
ax.set_xlabel('Crop Type')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy on Partial/Cropped Face Images\n(compared to full image baseline)')
ax.set_ylim(0, 1)
ax.axhline(y=val_acc, color='red', linestyle='--', label=f'Full Image Acc: {val_acc:.2f}')
ax.legend()

for bar, acc in zip(bars, accuracies):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
            f'{acc:.2f}', ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

print("\nSummary:")
print(f"  Full Image Accuracy:  {val_acc:.4f}")
for ct, acc in zip(crop_types, accuracies):
    print(f"  {ct:15s} Accuracy: {acc:.4f}  (drop: {val_acc - acc:+.4f})")
```

```python
# Visualize some partial image predictions
def visualize_crop_predictions(model, data_dir, class_names, crop_types=['eyes', 'left_half', 'right_half']):
    """Show side-by-side full vs cropped predictions."""
    
    # Pick one sample per class
    sample_images = []
    for class_name in class_names[:6]:  # Show 6 celebrities
        class_path = os.path.join(data_dir, class_name)
        img_name = os.listdir(class_path)[0]
        sample_images.append((class_name, os.path.join(class_path, img_name)))
    
    n_crops = len(crop_types) + 1  # +1 for original
    fig, axes = plt.subplots(len(sample_images), n_crops, figsize=(3 * n_crops, 4 * len(sample_images)))
    
    if len(sample_images) == 1:
        axes = axes.reshape(1, -1)
    
    col_labels = ['Original'] + crop_types
    
    for col in range(n_crops):
        axes[0][col].set_title(col_labels[col], fontsize=12, fontweight='bold')
    
    for row, (true_name, img_path) in enumerate(sample_images):
        img = load_and_preprocess_image(img_path)
        
        # Original
        input_orig = np.expand_dims(img, axis=0)
        probs_orig = model.predict(input_orig, verbose=0)[0]
        pred_orig = CLASS_NAMES[np.argmax(probs_orig)]
        
        axes[row][0].imshow(img)
        color = 'green' if pred_orig == true_name else 'red'
        axes[row][0].set_ylabel(true_name, fontsize=9, fontweight='bold')
        axes[row][0].set_xlabel(f"Pred: {pred_orig}\n{'✓' if pred_orig == true_name else '✗'}", 
                                fontsize=8, color=color)
        axes[row][0].tick_params(left=False, bottom=False, labelbottom=False, labelleft=False)
        
        # Cropped variants
        for col, crop_type in enumerate(crop_types, 1):
            cropped = crop_image(img, crop_type)
            cropped_resized = tf.image.resize(cropped, (IMG_SIZE, IMG_SIZE))
            input_crop = np.expand_dims(cropped_resized.numpy(), axis=0)
            
            probs_crop = model.predict(input_crop, verbose=0)[0]
            pred_crop = CLASS_NAMES[np.argmax(probs_crop)]
            
            axes[row][col].imshow(cropped)
            color = 'green' if pred_crop == true_name else 'red'
            axes[row][col].set_xlabel(f"Pred: {pred_crop}\n{'✓' if pred_crop == true_name else '✗'}", 
                                      fontsize=8, color=color)
            axes[row][col].tick_params(left=False, bottom=False, labelbottom=False, labelleft=False)
    
    plt.suptitle('Prediction on Full vs Cropped Images\n(green=correct, red=wrong)', fontsize=14)
    plt.tight_layout()
    plt.show()

visualize_crop_predictions(model, DATA_DIR, CLASS_NAMES)
```

## 6. Prediction Function — Drop Any Image In

```python
def predict_celebrity(img_path, model=model, class_names=CLASS_NAMES, 
                      show_img=True, top_k=5, crop_types=None):
    """
    Predict which celebrity is in an image.
    
    Parameters:
    -----------
    img_path : str
        Path to the input image.
    model : keras.Model
        Trained CNN model.
    class_names : list
        List of class names.
    show_img : bool
        Whether to display the image and prediction results.
    top_k : int
        Number of top predictions to show.
    crop_types : list or None
        If provided, also predict on cropped versions and show side-by-side.
        Options: 'eyes', 'lower', 'left_half', 'right_half', 'center'
    
    Returns
    -------
    dict with prediction results
    """
    img = load_and_preprocess_image(img_path)
    input_array = np.expand_dims(img, axis=0)
    
    probs = model.predict(input_array, verbose=0)[0]
    top_indices = np.argsort(probs)[::-1][:top_k]
    
    result = {
        'image_path': img_path,
        'top_predictions': [(class_names[i], probs[i]) for i in top_indices]
    }
    
    if show_img:
        if crop_types:
            n_cols = len(crop_types) + 1
            fig, axes = plt.subplots(1, n_cols, figsize=(3.5 * n_cols, 4))
            
            # Original
            axes[0].imshow(img)
            pred_name, conf = result['top_predictions'][0]
            axes[0].set_title(f'Original\n{pred_name}\nConf: {conf:.2%}', fontsize=10)
            axes[0].axis('off')
            
            # Crops
            for col, crop_type in enumerate(crop_types, 1):
                cropped = crop_image(img, crop_type)
                cropped_resized = tf.image.resize(cropped, (IMG_SIZE, IMG_SIZE))
                input_crop = np.expand_dims(cropped_resized.numpy(), axis=0)
                probs_crop = model.predict(input_crop, verbose=0)[0]
                top_idx = np.argmax(probs_crop)
                pred_crop = class_names[top_idx]
                conf_crop = probs_crop[top_idx]
                
                axes[col].imshow(cropped)
                axes[col].set_title(f'{crop_type}\n{pred_crop}\nConf: {conf_crop:.2%}', fontsize=10)
                axes[col].axis('off')
            
            plt.suptitle('Celebrity Guesser — Full vs Cropped', fontsize=13)
            plt.tight_layout()
            plt.show()
        else:
            fig, axes = plt.subplots(1, 2, figsize=(10, 4))
            
            # Show image
            axes[0].imshow(img)
            pred_name, conf = result['top_predictions'][0]
            axes[0].set_title(f'{pred_name}\nConfidence: {conf:.2%}', fontsize=12)
            axes[0].axis('off')
            
            # Show top predictions bar chart
            names = [p[0] for p in result['top_predictions']]
            confs = [p[1] for p in result['top_predictions']]
            colors = ['#2ecc71' if n == names[0] else '#3498db' for n in names]
            axes[1].barh(names[::-1], confs[::-1], color=colors[::-1])
            axes[1].set_xlabel('Confidence')
            axes[1].set_xlim(0, 1)
            axes[1].set_title('Top Predictions')
            
            for i, (name, c) in enumerate(zip(names[::-1], confs[::-1])):
                axes[1].text(c + 0.01, i, f'{c:.1%}', va='center', fontsize=9)
            
            plt.tight_layout()
            plt.show()
    
    # Print results
    print(f"\n{'='*50}")
    print(f"Image: {img_path}")
    print(f"\nTop {top_k} Predictions:")
    for i, (name, prob) in enumerate(result['top_predictions'], 1):
        bar = '█' * int(prob * 30)
        print(f"  {i}. {name:<25s} {prob:6.2%}  {bar}")
    print(f"{'='*50}\n")
    
    return result

# Example usage:
# result = predict_celebrity('path/to/any/image.jpg')
# result = predict_celebrity('path/to/any/image.jpg', crop_types=['eyes', 'left_half'])
print("predict_celebrity() function ready!")
```

```python
# Quick demo — test on a few random images from the dataset
demo_images = []
for class_name in CLASS_NAMES[:3]:
    class_path = os.path.join(DATA_DIR, class_name)
    images = os.listdir(class_path)
    # Pick 2 random images (different from first one used in visualization)
    for img_name in images[5:7]:
        demo_images.append((class_name, os.path.join(class_path, img_name)))

for true_name, img_path in demo_images:
    print(f"[TRUE: {true_name}]")
    predict_celebrity(img_path, show_img=True, crop_types=['eyes', 'left_half'])
```

## Summary

This notebook implements a **Celebrity Guesser CNN** with the following components:

| Component | Details |
|---|---|
| **Dataset** | 17 celebrities × 100 images = 1,700 total |
| **Image Size** | 64×64 pixels |
| **Architecture** | 3 Conv blocks (32→64→128 filters) + Dense head |
| **Augmentation** | Rotation, shift, flip, zoom |
| **Training** | Adam optimizer, early stopping, LR reduction |
| **Evaluation** | Accuracy, confusion matrix, classification report |
| **Partial Image Test** | Eyes, lower face, left/right halves, center crop |
| **Prediction** | `predict_celebrity(img_path)` function for any image |

**Key Insight:** The model achieves high accuracy on full-face images but **significantly drops** on cropped/partial inputs — this is intentional to demonstrate CNN limitations and the importance of data diversity during training.
